```python
import pandas as pd
```

```python
authors = pd.DataFrame({'author_id': [1, 2, 3], 'author_name': ['Тургенев', 'Чехов', 'Островский']})
```

```python
book_data = {'author_id': [1, 1, 1, 2, 2, 3, 3], 'book_title': 
             ['Отцы и дети', 'Рудин', 'Дворянское гнездо', 'Толстый и тонкий', 'Дама с собачкой', 'Гроза', 
              'Таланты и поклонники'], 'price': [450, 300, 350, 500, 450, 370, 290]}
```

```python
book = pd.DataFrame(book_data)
book
```

```python
authors
```

```python
authors_price = pd.merge(authors, book, on='author_id', how='inner')
authors_price
```

```python
top5 = authors_price.nlargest(5, 'price')
top5
```

```python
authors_groupby = authors_price.groupby('author_name')
max_price = pd.DataFrame(authors_groupby.agg({'price': 'max'}))
min_price = pd.DataFrame(authors_groupby.agg({'price': 'min'}))
mean_price = pd.DataFrame(authors_groupby.agg({'price': 'mean'}))
authors_im = pd.merge(max_price, min_price, on='author_name', how='left')
authors_stat = pd.merge(authors_im, mean_price, on='author_name', how='left')
authors_stat
```

```python
authors_stat.rename(columns = {'price_x' : 'max_price', 'price_y' : 'min_price', 'price' : 'mean_price' }, inplace=True)
authors_stat
```
